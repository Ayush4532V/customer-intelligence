```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# load cleaned data
df = pd.read_csv("../data/cleaned.csv")

df.head()
plt.figure(figsize=(14,10))
corr = df.corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5
)

plt.title("Correlation Heatmap of Customer Features")
plt.show()



```

```python
spend_cols = [
    'MntWines','MntFruits','MntMeatProducts',
    'MntFishProducts','MntSweetProducts','MntGoldProds'
]

df['Total_Spending'] = df[spend_cols].sum(axis=1)

df[['Total_Spending']].head()
plt.figure(figsize=(8,5))
sns.histplot(df['Total_Spending'], bins=30, kde=True)

plt.title("Distribution of Total Customer Spending")
plt.xlabel("Total Spending")
plt.ylabel("Customer Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=df['Age'],
    y=df['Total_Spending'],
    alpha=0.6
)

plt.title("Age vs Total Spending")
plt.xlabel("Age")
plt.ylabel("Total Spending")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=df['Income'],
    y=df['Total_Spending'],
    alpha=0.6
)

plt.title("Income vs Total Spending")
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.show()
```

```python
channels = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

df[channels].mean().plot(kind='bar', figsize=(8,5))

plt.title("Average Purchases by Channel")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['Response'])

plt.title("Customer Response to Marketing Campaign")
plt.xlabel("Response (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

Key Insights from EDA:

Income and total spending show a positive correlation

Majority of customers fall into lowmedium spending range

Web purchases dominate over catalog purchases

Marketing campaign response is imbalanced
